```python
import urllib.request
import urllib.parse

from bs4 import BeautifulSoup

def save_html_to_file(html, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(html)
```

```python
seed_url = 'https://press.un.org/en'

urls = [seed_url]
seen_url=[seed_url]
result_urls = []
press_num = 0
min_num_urls = 10   


while press_num < min_num_urls and len(urls)>0:
        
        try:
            current_url = urls.pop(0)
            req = urllib.request.Request(current_url, headers={'User-Agent': 'Mozilla/5.0'})
            webpage = urllib.request.urlopen(req).read()
            soup = BeautifulSoup(webpage, features='html.parser')
            tag = soup.find('a', href="/en/press-release", text='Press Release')
            
            
            if tag:
                text = soup.get_text()
                if 'crisis' in text.lower():
                    press_num += 1
                    #print('press_num = '+ str(press_num))
                    result_urls.append(current_url)
                    save_html_to_file(str(soup), f'1_{press_num}.txt')
            #else:
                #print('None')
        except:
            continue
            

        for a_tag in soup.find_all('a', href=True):
            
            org_child_url = a_tag.get('href')
            child_url = urllib.parse.urljoin(seed_url, org_child_url)
            
            if child_url not in seen_url and seed_url in child_url:
            
                seen_url.append(child_url)
                urls.append(child_url)
print('10 United Nations press releases containing the word “crisis” :')            
for results in result_urls:
    print(results)
```
